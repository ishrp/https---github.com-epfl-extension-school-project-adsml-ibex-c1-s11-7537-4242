```python
# 4. Groupping the communes by canton and saving them into separate .csv files
cantons=df["Canton"].unique() #getting the canton's names 

for i in range(len(cantons)):
    new_filename = str(cantons[i])+str(".scv")
    df.to_csv(new_filename, encoding="utf-8") #saving the csv file
    
```

```python
# 5. Computing the population density at the canton level and ranking the cantons from most dense to least dense 
#First we will need to calculate the number of residents and the total surface in every canton 
grouped_df = df.groupby("Canton").agg(
    {
        "Residents" : "sum",
        "Surface area in km²" : "sum", 
    }
) 

#Then we create the new column with the population density of each canton 
grouped_df["Population density"] = grouped_df["Residents"] /grouped_df["Surface area in km²"] 
grouped_df.sort_values("Population density", ascending=False) 

#Although the most populated canton in terms of residents is Zurich, it is not the most densily populated. Due to its bigger size, it is ranked only as the third mostly densed canton ceding to Basel and Geneva. 
#Commune Basel having 7 times less of population that Zurich, but with its very small surface appears to be the most densily populated canton 
```

```python
# 6. Computing the number of communes in each canton where more than 50 percent of their populations are aged between 20 and 64 years old.

df_2064 = df[df["20-64 years"] > 50] #creating the separate dataframe which satisfies the condition stated in the task 

agg_fnct = {"Commune" : 'count'}
group_df = df_2064.groupby("Canton").agg(agg_fnct) #Computing how many communes satisfy the condition stated in the task 

group_df
```

```python
# 7. Computng the difference between the maximum and minimum elevations for each canton. Find the top 5 cantons that have the largest range of elevations

agg_fnct = {
    "Elevation" : lambda x: max(x) - min(x)
}
group_df = df_2064.groupby("Canton").agg(agg_fnct) #identifying  the difference between the min and max elevation for each canton 

group_df.sort_values("Elevation", ascending = False).head(5) #finding the top 5 cantons with largest range of elevation 
```

```python
# C. Exploration : visualizations 
#1 Obtaining a horizontal bar plot that shows the top 10 populated communes.

df_pop = df.sort_values(by = "Residents", ascending=False)

density = df_pop[0:10]["Residents"]
commune = df_pop[0:10]["Commune"]
fig = plt.figure(figsize=(10, 5))
plt.barh(commune, density, color='maroon')
 
plt.xlabel("Residents")
plt.ylabel("Commune")
plt.title("The top 10 populated communes")
plt.show()
```

```python
# 2. For the top 10 populated communes of the previous step, your task now is to plot a horizontal stacked bar chart that shows how their lands are divided into the 4 area types

settlement = agriculture = df_pop[0:10]["Settlement area"]
agriculture = df_pop[0:10]["Agricultural area"]
wood = df_pop[0:10]["Wooded area"]
unproductive = df_pop[0:10]["Unproductive area"]
commune = df_pop[0:10]["Commune"]

fig = plt.figure()
plt.barh(commune, settlement, color='blue', label = "Settelment area")
plt.barh(commune, agriculture, color='orange', left = settlement, label = "Agriculture area")
plt.barh(commune, wood, color='green', left = settlement + agriculture, label = "Wooden area")
plt.barh(commune, unproductive, color='red', left = settlement + agriculture+wood, label = "Unproductive area")
plt.xlabel("area type in %")
plt.ylabel("Commune")
plt.title("Top 10 populated communes by type of area")
plt.legend(loc = "right", bbox_to_anchor = (1.4, 0.9))
plt.show()
#There is no clear relationship betweeen the type of the area and the population of the commune. We could have said that the most populated areas tend to be settlement areas, but we have an outlier with Lugano. 
```

```python
# E. Matrices 
# Define a data frame matrix whose rows correspond to communes and the columns to the cantons. Fill in the matrix with 0/1 values where entry (i,j) is a 1 if the commune in row i is in the canton in column j and a 0 otherwise.

nb_cantons = df["Canton"].nunique() #extracting the number of cantons 
nb_communes = df["Commune"].nunique() #exctracting the number of communes 

cantons = list(df["Canton"].unique()) #extracting the list of cantons
communes = list(df["Commune"].unique()) #extracting the list of communes 

A = np.zeros((nb_communes, nb_cantons)) #creating the matrix with rows corresponding to communes, and columns correspoding to cantons (in terms of dimension) with zeros everywhere 

#filling in the matrix with values of 1 if a given commune is located in the canton in question 
for j in range(nb_cantons): 
    for i in range(nb_communes): 
        if df["Canton"][i] == cantons[j]: 
            A[i, j] = 1
```
